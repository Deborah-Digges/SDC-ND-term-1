```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
image = mpimg.imread('./test.jpg')
plt.imshow(image);
```

```python
print("The image is: ", type(image), " with dimensions", image.shape)
```

```python
ysize = image.shape[0]
xsize = image.shape[1]
```

```python
# copy the image
color_select = np.copy(image)
```

```python
# Define our color selection criteria
# color threshold in the variables red_threshold, green_threshold, 
# and blue_threshold and populate rgb_threshold with these values. 
# This vector contains the minimum values for red, green, and blue (R,G,B) that I will allow in my selection.
red_threshold = 0
green_threshold = 0
blue_threshold = 0
rgb_threshold = [red_threshold, green_threshold, blue_threshold]
```

```python
# Use a "bitwise OR" to identify pixels below the threshold
thresholds = (image[:,:,0] < rgb_threshold[0]) \
            | (image[:,:,1] < rgb_threshold[1]) \
            | (image[:,:,2] < rgb_threshold[2])
```

```python
# Since we have set thresholds of (0,0,0), no pixels have RGB components below this value
np.any(thresholds)
```

```python
color_select[thresholds] = [0,0,0]
```

```python
plt.imshow(color_select);
```
